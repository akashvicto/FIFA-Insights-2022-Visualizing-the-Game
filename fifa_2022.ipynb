### **Total No of goals per team**

```python
a=df.groupby('team1')['number of goals team1'].sum()
b=df.groupby('team2')['number of goals team2'].sum()
total=a+b
total=total.sort_values(ascending=False)
total.plot(kind='bar', color='skyblue', figsize=(10, 6))

# Add labels and a title
plt.xlabel('Team')
plt.ylabel('Total Goals')
plt.title('Total Goals Scored by Each Team')

# Show the plot
plt.show()
```

### **Goals inside the penalty area**

```python
in1=df.groupby('team1')['goal inside the penalty area team1'].sum()
in2=df.groupby('team2')['goal inside the penalty area team2'].sum()
inside=in1+in2
inside=inside.sort_values(ascending=False)
inside.plot(kind='bar',color='black',figsize=(10,6))

# Add labels and a title
plt.xlabel('Team')
plt.ylabel('Goals')
plt.title('Goals Scored inside the penalty area')

# Show the plot
plt.show()
```

### **Goals outside the penalty area**

```python
out1=df.groupby('team1')['goal outside the penalty area team1'].sum()
out2=df.groupby('team2')['goal outside the penalty area team2'].sum()
outside=out1+out2
outside=outside.sort_values(ascending=False)
outside.plot(kind='bar',color='black',figsize=(10,6))

# Add labels and a title
plt.xlabel('Team')
plt.ylabel('Goals')
plt.title('Goals Scored outside the penalty area')

# Show the plot
plt.show()
```

### **Total assists per team**

```python
ass1=df.groupby('team1')['assists team1'].sum()
ass2=df.groupby('team2')['assists team2'].sum()
assist=ass1+ass2
assist.plot(kind='bar',color='black',figsize=(10,6))

# Add labels and a title
plt.xlabel('Team')
plt.ylabel('Assists')
plt.title('Total assists per team')

# Show the plot
plt.show()
```

### **Average attempts per team**

```python
df[['team1', 'team2','total attempts team2','total attempts team1','on target attempts team1',
'on target attempts team2','off target attempts team1', 'off target attempts team2']]
```

```python
att1=df.groupby('team1')['total attempts team1'].mean()
att2=df.groupby('team2')['total attempts team2'].mean()
attempts=(att1+att2)/2

# Create a pie chart
plt.figure(figsize=(12, 12))
plt.pie(attempts, labels=attempts.index, autopct='%1.1f%%', startangle=140)
plt.title('Average Total Attempts by Team')

# Display the pie chart
plt.show()
```

### **Passes complited per team**

```python
pass1 = df.groupby('team1')['passes team1'].sum()
pass2 = df.groupby('team2')['passes team2'].sum()
total_passes = pass1.add(pass2, fill_value=0)


pass_com1=df.groupby('team1')['passes completed team1'].sum()
pass_com2=df.groupby('team2')['passes completed team2'].sum()
total_pass_com=pass_com1.add(pass_com2,fill_value=0)


teams = total_passes.index
bar_width = 0.3
index = np.arange(len(teams))

# Create a figure and axis with a specified size
fig, ax = plt.subplots(figsize=(15, 15))

# Plot the bars for total passes and total passes completed
bar1 = ax.barh(index, total_passes, bar_width, label='Total Passes', color='skyblue')
bar2 = ax.barh(index + bar_width, total_pass_com, bar_width, label='Total Passes Completed', color='lightcoral')

# Set y-axis labels and titles
ax.set_yticks(index + bar_width / 2)
ax.set_yticklabels(teams)
ax.set_xlabel('Passes')
ax.set_ylabel('Teams')
ax.set_title('Passes statics by Team')

# Display legend
ax.legend()

# Show the plot
plt.show()
```

### **Fouls**

```python
fou1=df.groupby('team1')['fouls against team1'].sum()
fou2=df.groupby('team2')['fouls against team2'].sum()
total_fouls=fou1+fou2
total_fouls.plot(kind='bar', color='skyblue',figsize=(10,10))

# Add labels and a title
plt.xlabel('Team')
plt.ylabel('Fouls')
plt.title('Total no of fouls aganist Each Team')

# Show the plot
plt.show()
```

### **Yellow / Red Cards**

```python

yel1=df.groupby('team1')['yellow cards team1'].sum()
yel2=df.groupby('team2')['yellow cards team2'].sum()
yellow=yel1+yel2



red1=df.groupby('team1')['red cards team1'].sum()
red2=df.groupby('team2')['red cards team2'].sum()
red=red1+red2

teams = yellow.index
bar_width = 0.3
index =np.arange(len(teams))

fig, ax = plt.subplots(figsize=(15, 15))


bar1 = ax.barh(index, yellow, bar_width, label='Yellow cards', color='yellow')
bar2 = ax.barh(index + bar_width, red, bar_width, label='Red cards', color="Red")


ax.set_yticks(index + bar_width / 2)
ax.set_yticklabels(teams)
ax.set_xlabel('Yellow/Red cards')
ax.set_ylabel('Teams')
ax.set_title('Cards by Team')

# Display legend
ax.legend()

# Show the plot
plt.show()
```

### **Total crosses per Team**

```python
cro1=df.groupby('team1')['crosses team1'].sum()
cro2=df.groupby('team2')['crosses team2'].sum()
cross=cro1+cro2



cro_com1=df.groupby('team1')['crosses completed team1'].sum()
cro_com2=df.groupby('team2')['crosses completed team2'].sum()
cro_completed=cro_com1+cro_com2

teams = cross.index
bar_width = 0.3
index = np.arange(len(teams))


fig, ax = plt.subplots(figsize=(20, 20))


bar1 = ax.barh(index, cross, bar_width, label='Total Crosses', color='red')
bar2 = ax.barh(index + bar_width, cro_completed, bar_width, label='cross_completed', color="blue")


ax.set_yticks(index + bar_width/2)
ax.set_yticklabels(teams)
ax.set_xlabel('Total crosses')
ax.set_ylabel('Teams')
ax.set_title('Total crosses per team')

# Display legend
ax.legend()

# Show the plot
plt.show()
```

### **defensive pressures applied team by Groups**

```python
pos1 = df.groupby(['team1', 'category'])['defensive pressures applied team1'].sum()
pos2 = df.groupby(['team2', 'category'])['defensive pressures applied team2'].sum()

# Combine the totals
possession = (pos1 + pos2)

# Get unique categories
groups = ['Group A', 'Group B', 'Group C', 'Group D', 'Group E', 'Group F', 'Group G', 'Group H']

# Create a separate bar chart for each unique category
for group in groups:
    group_data = possession.xs(group, level='category', drop_level=False)

    # Prepare data for the bar chart
    teams = group_data.index.get_level_values('team1')
    values = group_data.values  # Use the actual possession values

    # Create the bar chart
    plt.figure(figsize=(8, 6))
    plt.bar(teams, values, width=0.3, color='green')
    plt.xlabel('Teams')
    plt.ylabel('Defensive pressure')
    plt.title(f'defensive pressures applied team - {group}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()

    # Display the bar chart
    plt.show()
```
